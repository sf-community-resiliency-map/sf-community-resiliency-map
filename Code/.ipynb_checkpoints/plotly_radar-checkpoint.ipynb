```python
import pandas as pd
import numpy as np

df = pd.read_csv('../Data/Community_Resiliency_Indicator_System.csv')
df = df[['Neighborhood', 'Haz_Score', 'Env_Score', 'Trans_Score', 'Com_Score', 'PR_Score',
     'House_Score', 'Ec_Score', 'Health_Score', 'Dem_Score', 'Res_Score']].dropna()
```

```python
{
type: 'scatterpolar',
r: [39, 28, 8, 7, 28, 39],
theta: ['A','B','C', 'D', 'E', 'A'],
fill: 'toself',
name: 'Group A'
},
{
type: 'scatterpolar',
r: [1.5, 10, 39, 31, 15, 1.5],
theta: ['A','B','C', 'D', 'E', 'A'],
fill: 'toself',
name: 'Group B'
}
```

```python
i = 0
for index, row in df.iterrows():
    print("{")
    print("type: 'scatterpolar',")
    print("r: %s," % list(row[1:]))
    print("theta: ['Haz_Score', 'Env_Score', 'Trans_Score', 'Com_Score', 'PR_Score', 'House_Score', 'Ec_Score', 'Health_Score', 'Dem_Score', 'Res_Score'],")
    print("fill: 'toself',")
    print("name: '%s'," % row[0])
    if( int(i/6) > 0):
        print('subplot: "polar%d"' % (int(i/6)+1))
    print("},")
    i += 1
```
