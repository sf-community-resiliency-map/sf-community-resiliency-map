```python
i = 1
for s in ['Haz_Score', 'Env_Score', 'Trans_Score', 'Com_Score', 'PR_Score',
     'House_Score', 'Ec_Score', 'Health_Score', 'Dem_Score', 'Res_Score']:
    print('geo_source = GeoJSONDataSource(geojson=json.dumps(jo))')
    print('TOOLS = "pan,wheel_zoom,box_zoom,reset,hover,save"')
    print("color_mapper = LinearColorMapper(palette='Blues6', low=5, high=1)")
    print("p%d = figure(title='%s Choropleth Map', tools=TOOLS, x_axis_location=None, y_axis_location=None, width=1000, height=600)" %(i, name_dict[s]))
    print("p%d.grid.grid_line_color = None" %i)
    print("p%d.patches('xs', 'ys', fill_alpha=0.7, fill_color={'field':'%s', 'transform':color_mapper}, line_color='white', line_width=0.5, source=geo_source)" %(i, s))
    print("hover = p%d.select_one(HoverTool)" %i)
    print('hover.point_policy = "follow_mouse"')
    print('hover.tooltips = [("Name:", "@name"), ("%s:", "@%s")]' %(s, s))
    print('tab%d = Panel(child=p%d, title="%s")' %(i, i, name_dict[s]))
    print()
   
    i += 1
```
